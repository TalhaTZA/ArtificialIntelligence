```python
train_data=data[0:108,:-1].astype('float32')
train_labels=data[0:108,-1].astype('float32')
test_data=data[108:,:-1].astype('float32')
test_labels=data[108:,-1].astype('float32')
print(train_data.shape)
print(train_labels.shape)
print(test_data.shape)
print(test_labels.shape)
```

```python
train_labels
```

```python
from keras import models
from keras import layers
def build_model():
  model = models.Sequential()
  model.add(layers.Dense(64, activation='relu',
  input_shape=(train_data.shape[1],)))
  model.add(layers.Dense(64, activation='relu'))
  model.add(layers.Dense(1,activation='sigmoid'))
  model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
  return model
```

```python
import numpy as np
k=4
num_val_samples = len(train_data) // k
num_epochs = 50
all_scores = []
for i in range(k):
  print('processing fold #', i)
  val_data = train_data[i * num_val_samples: (i + 1) * num_val_samples]
  val_targets = train_labels[i * num_val_samples: (i + 1) * num_val_samples]
  partial_train_data = np.concatenate(
  [train_data[:i * num_val_samples],
  train_data[(i + 1) * num_val_samples:]],
  axis=0)
  partial_train_targets = np.concatenate(
  [train_labels[:i * num_val_samples],
  train_labels[(i + 1) * num_val_samples:]],
  axis=0)
  model = build_model()
  model.fit(partial_train_data, partial_train_targets,
  epochs=num_epochs, batch_size=25, verbose=1)
  val_loss, val_acc = model.evaluate(val_data, val_targets, verbose=1)
  check=model.evaluate(val_data, val_targets, verbose=1)
  all_scores.append(val_acc)
```

```python
num_epochs = 50
all_val_acc_histories = []
for i in range(k):
  print('processing fold #', i)
  val_data = train_data[i * num_val_samples: (i + 1) * num_val_samples]
  val_targets = train_labels[i * num_val_samples: (i + 1) * num_val_samples]
  partial_train_data = np.concatenate(
  [train_data[:i * num_val_samples],
  train_data[(i + 1) * num_val_samples:]],
  axis=0)
  partial_train_targets = np.concatenate(
  [train_labels[:i * num_val_samples],
  train_labels[(i + 1) * num_val_samples:]],
  axis=0)
  model = build_model()
  history = model.fit(partial_train_data, partial_train_targets,
  validation_data=(val_data, val_targets),
  epochs=num_epochs, batch_size=5, verbose=1)
  val_acc_history = history.history['val_acc']
  all_val_acc_histories.append(val_acc_history)
```

```python
type(all_val_acc_histories)
```

```python
#taking mean of each column
average_val_acc_history = [
np.mean([x[i] for x in all_val_acc_histories]) for i in range(num_epochs)]
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, len(average_val_acc_history) + 1), average_val_acc_history)
plt.xlabel('Epochs')
plt.ylabel('Validation accuracy')
plt.show()
```
