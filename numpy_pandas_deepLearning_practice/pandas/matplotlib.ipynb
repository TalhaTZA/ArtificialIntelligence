```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=np.arange(10)
```

```python
plt.plot(data)
plt.show()
plt.gcf().clear()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(),'k--')
plt.show()
plt.gcf().clear()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)
ax1.plot(np.arange(10))
ax2.scatter(np.arange(10),np.arange(10,20))
ax3.bar([1,2],[30.0,40.0])
plt.show()
plt.gcf().clear()
```

```python
plt.plot(np.random.randn(30).cumsum(),'go--',label='test') #green o -- line
#plt.xticks([1,2,3])
plt.title("graph of student vs marks")
plt.legend(loc='best')
plt.show()
plt.gcf().clear()
```
